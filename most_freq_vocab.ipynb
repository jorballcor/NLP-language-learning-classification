```python
import spacy
```

# FR

```python
nlp = spacy.load("fr_core_news_sm")
```

```python
with open("fr_10k_clean.txt", "r", encoding="utf-8") as f:
    mots = [line.strip() for line in f if line.strip()]
    
```

```python
len(mots)
```

```python
data = []

for doc in nlp.pipe(mots):
    for token in doc:
        gender = token.morph.get("Gender")
        number = token.morph.get("Number")
        data.append({
            'Mot': token.text,
            'POS': token.pos_,
            'Gender': gender[0] if gender else None,
            'Number': number[0] if number else None
        })
```

```python
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
df.to_excel("fr_10k.xlsm", index=False)
```
